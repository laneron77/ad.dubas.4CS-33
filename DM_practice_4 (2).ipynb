#### **Система Лоренца:**

Система Лоренца описується трьома диференціальними рівняннями:

$$
\begin{cases}
\frac{dx}{dt} = \sigma (y - x) \\
\frac{dy}{dt} = x (\rho - z) - y \\
\frac{dz}{dt} = xy - \beta z
\end{cases}
$$

де:
- $( \sigma )$ — коефіцієнт Прандтля (зазвичай $ \sigma = 10 $),
- $( \rho )$ — числовий параметр для турбулентності (зазвичай $ \rho = 28 $),
- $( \beta )$ — коефіцієнт відношення висоти до довжини (зазвичай $ \beta = 8/3 $).

#### **Об'єкти Нарендра-Пархтізаратхі:**

Об'єкти Нарендра-Пархтізаратхі описуються наступними рівняннями:

$$
\begin{cases}
x_1(t+1) = \alpha_1 x_1(t) + \beta_1 x_2(t) + \gamma_1 x_3(t) + \delta_1 \\
x_2(t+1) = \alpha_2 x_1(t) + \beta_2 x_2(t) + \gamma_2 x_3(t) + \delta_2 \\
x_3(t+1) = \alpha_3 x_1(t) + \beta_3 x_2(t) + \gamma_3 x_3(t) + \delta_3
\end{cases}
$$

де $(\alpha_i)$, $(\beta_i)$, $(\gamma_i)$ і $(\delta_i)$ — коефіцієнти для кожного з рівнянь.

```python
# Параметри для системи Лоренца
sigma = 10.0
rho = 28.0
beta = 8.0 / 3.0

# Крок часу
dt = 0.001

# Кількість точок даних
num_points = 10000

# Ініціалізація масивів для x, y, z
x = np.zeros(num_points)
y = np.zeros(num_points)
z = np.zeros(num_points)

# Початкові умови
x[0], y[0], z[0] = 1.0, 1.0, 1.0

# Чисельне рішення з використанням методу Ейлера
for i in range(1, num_points):
    dx = sigma * (y[i-1] - x[i-1]) * dt
    dy = (x[i-1] * (rho - z[i-1]) - y[i-1]) * dt
    dz = (x[i-1] * y[i-1] - beta * z[i-1]) * dt
    x[i] = x[i-1] + dx
    y[i] = y[i-1] + dy
    z[i] = z[i-1] + dz

# Збереження даних у DataFrame
lorenz_data = pd.DataFrame({'x': x, 'y': y, 'z': z})

# Візуалізація
plt.figure(figsize=(10, 6))
plt.plot(x, label='x')
plt.plot(y, label='y')
plt.plot(z, label='z')
plt.legend()
plt.title('Lorenz System')
plt.xlabel('Time steps')
plt.ylabel('Values')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Функція для генерації даних для об'єктів Нарендра-Пархтізаратхі
def narendra_parthasarathy_system(num_points):
    # Ініціалізація масивів для x1, x2, x3
    x1 = np.zeros(num_points)
    x2 = np.zeros(num_points)
    x3 = np.zeros(num_points)

    # Випадкові коефіцієнти з обмеженим діапазоном
    alpha1, beta1, gamma1, delta1 = np.random.uniform(0, 0.5, 4)
    alpha2, beta2, gamma2, delta2 = np.random.uniform(0, 0.5, 4)
    alpha3, beta3, gamma3, delta3 = np.random.uniform(0, 0.5, 4)

    # Початкові умови
    x1[0], x2[0], x3[0] = 1.0, 1.0, 1.0

    # Генерація даних
    for t in range(1, num_points):
        # Перевірка на нуль у знаменнику для x1
        sin_x1 = np.sin(x1[t-1])
        if sin_x1 != 0:
            x1[t] = alpha1 * x1[t-1] - beta1 * x2[t-1] / sin_x1
        else:
            alpha1 * x1[t-1] + beta1 * x2[t-1] + gamma1 * x3[t-1] + delta1

        # Перевірка на нуль у знаменнику для x2
        cos_x1 = np.cos(x1[t-1])
        if cos_x1 != 0:
            x2[t] = -alpha2 * x2[t-1] + beta2 / cos_x1
        else:
            x2[t] = alpha2 * x1[t-1] + beta2 * x2[t-1] + gamma2 * x3[t-1] + delta2

        # Обчислення x3
        x3[t] = alpha3 * x1[t-1] + beta3 * x2[t-1] + gamma3 * x3[t-1] + delta3
        
        # Перевірка на переповнення і скидання у разі необхідності
        if any(np.isinf([x1[t], x2[t], x3[t]])) or any(np.isnan([x1[t], x2[t], x3[t]])):
            x1[t], x2[t], x3[t] = 0, 0, 0  # Скидання для уникнення переповнення

    return x1, x2, x3

# Кількість точок даних
num_points = 10000

# Генерація даних
x1, x2, x3 = narendra_parthasarathy_system(num_points)

# Збереження даних у DataFrame
narendra_data = pd.DataFrame({'x1': x1, 'x2': x2, 'x3': x3})

# Візуалізація
plt.figure(figsize=(10, 6))
plt.plot(x1, label='x1', color='blue')
plt.plot(x2, label='x2', color='orange')
plt.plot(x3, label='x3', color='green')
plt.legend()
plt.title('Narendra-Parthasarathy Objects')
plt.xlabel('Time steps')
plt.ylabel('Values')
plt.grid()
plt.show()
```
